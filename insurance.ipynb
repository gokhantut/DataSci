```python
# make environment
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
# Import the data from the csv file
data = pd.read_csv('insurance.csv')

# Extract the relevant columns
age = data['age']
charges = data['charges']

# Plot a scatter plot with age on the x-axis and charges on the y-axis
plt.scatter(age, charges)
plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('Scatter plot of age vs charges')
plt.show()

# Use linear regression to fit a model to the data and make predictions
X = age.values.reshape(-1, 1)
y = charges.values.reshape(-1, 1)
regressor = LinearRegression()
regressor.fit(X, y)
y_pred = regressor.predict(X)

# Plot the best-fit line on top of the scatter plot
plt.scatter(age, charges)
plt.plot(age, y_pred, color='red')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('Age vs Charges with Best-fit line')
plt.show()

# Take user input for new data and make a prediction
while True:
    input_age = input('Enter age or "exit" to quit: ')
    if input_age == 'exit':
        print('Exiting the program.')
        break
    else:
        X_new = [[float(input_age)]]
        y_new = regressor.predict(X_new)
        print('Predicted charges for age {}: ${:.2f}'.format(input_age, y_new[0][0]))
```
